## Lists

A `list` is an ordered, changeable ("mutable") collection, written with square brackets.
Items are accessed by a zero-based index (`my_list[0]` is the first item). Negative indices
count from the end (`my_list[-1]` is the last item).

```python
colors = ["red", "green", "blue"]

print(colors[0])
print(colors[-1])
print(len(colors))

colors.append("yellow")
print(colors)
```

## Slicing

Slicing (`my_list[start:stop]`) gets a sub-section of a list — `start` is included, `stop` is
not. Leaving `start` or `stop` empty means "from the beginning" or "to the end".

```python
letters = ["a", "b", "c", "d", "e"]

print(letters[1:3])   # b, c
print(letters[:2])    # a, b
print(letters[3:])    # d, e
```

## Tuples

A `tuple` is like a list but immutable — once created, it can't be changed. Use tuples for
fixed groups of values that shouldn't be modified, like coordinates. Written with
parentheses (or often no brackets at all).

```python
point = (3, 4)
print(point[0], point[1])

# This would raise an error if uncommented, because tuples can't be changed:
# point[0] = 99
```

## Dictionaries

A `dict` maps keys to values — like a lookup table. Written with curly braces and `key:
value` pairs. Keys are typically strings but can be any immutable type. Access a value with
`my_dict[key]`, and check whether a key exists with `in`.

```python
ages = {"Ada": 36, "Alan": 41, "Grace": 85}

print(ages["Ada"])
print("Alan" in ages)
ages["Linus"] = 54
print(ages)
```

## Sets

A `set` is an unordered collection of unique values — duplicates are automatically removed.
Sets are useful for membership checks and removing duplicates from a list. Written with
curly braces, or `set()` for an empty one (`{}` alone makes an empty dict, not a set).

```python
unique_visitors = {"alice", "bob", "alice", "carol"}
print(unique_visitors)   # "alice" only appears once
print(len(unique_visitors))
```

## Looping over dicts

Use `.items()` to loop over a dict's keys and values together — this is the idiomatic way,
rather than looking up each value by key inside the loop.

```python
scores = {"Ada": 95, "Alan": 88, "Grace": 91}

for name, score in scores.items(): # name and scores are the 2 variables created
    print(f"{name}: {score}")
```

## Recap

- `list` — ordered, mutable, indexable with `[]`, sliceable with `[start:stop]`.
- `tuple` — ordered, immutable — for fixed groups of values.
- `dict` — key-value mapping; loop with `.items()`.
- `set` — unordered, unique values only; great for de-duplication and membership checks.
