## Strings are sequences

A string behaves like a sequence of characters — you can index and slice it just like a
list. Strings are also immutable: methods that "change" a string actually return a *new*
string, they never modify the original in place.

```python
word = "python"

print(word[0])
print(word[-1])
print(word[0:3])
print(len(word))
```

## Case and whitespace methods

Common string methods: `.upper()`, `.lower()`, `.title()`, and `.strip()` (removes leading/
trailing whitespace). Remember: none of these change the original string — they return a new
one, so you need to assign or use the result.

```python
messy = "   Hello, World!   "

print(messy.strip())
print(messy.strip().upper())
print(messy.strip().lower())
```

## Splitting and joining

`.split(sep)` breaks a string into a list of pieces (splits on whitespace if `sep` is
omitted). `sep.join(list)` does the reverse: glues a list of strings together with `sep`
between each one.

```python
csv_row = "Ada,36,Programmer"
fields = csv_row.split(",")
print(fields)

rejoined = " | ".join(fields)
print(rejoined)
```

## Searching and replacing

`.replace(old, new)` swaps every occurrence of `old` with `new`. The `in` operator checks
whether a substring exists. `.startswith()` / `.endswith()` check the ends specifically.

```python
sentence = "I like cats and cats like me"

print("cats" in sentence) # checks for cats in sentence and returns true value
print(sentence.replace("cats", "dogs")) # changes cats for dogs
print(sentence.startswith("I like"))
```

## f-strings revisited: format specs

You saw f-strings in [[01_variables_and_types]]. They also support format specs after a
colon, for things like decimal precision: `f"{value:.2f}"` rounds a float to 2 decimal
places.

```python
pi = 3.14159265

print(f"{pi:.2f}")
print(f"{pi:.4f}")
```

## Recap

- Strings are indexable/sliceable sequences, but immutable — methods return new strings.
- `.strip()`, `.upper()`, `.lower()`, `.title()` handle case and whitespace.
- `.split()` / `"sep".join()` convert between strings and lists of strings.
- `.replace()`, `in`, `.startswith()`, `.endswith()` search and transform text.
- f-string format specs (`f"{x:.2f}"`) control number formatting inline.
