## Variables and assignment

A variable is a name that points to a value. Python figures out the type from the value
itself — you don't declare it up front like in some other languages. Use `snake_case` for
variable names (all lowercase, words separated by underscores) — this is the standard Python
convention (see [PEP 8](https://peps.python.org/pep-0008/)), and code that breaks it reads as
unusual to other Python programmers.

Assignment uses `=`. You can check a variable's type with the built-in `type()` function.

```python
student_name = "Ada"
student_age = 28

print(student_name)
print(type(student_name))
print(student_age)
print(type(student_age))
```

## Core built-in types

Python has a handful of built-in types you'll use constantly:

| Type    | Example       | Meaning                          |
|---------|---------------|-----------------------------------|
| `int`   | `42`          | whole number                     |
| `float` | `3.14`        | number with a decimal point      |
| `str`   | `"hello"`     | text, in quotes                  |
| `bool`  | `True`/`False`| a truth value                    |

Notice `bool` values are capitalized — `True` and `False`, not `true`/`false`. This trips up
people coming from other languages.

```python
price = 19.99
in_stock = True

print(type(price))
print(type(in_stock))
```

## Arithmetic operators

Python supports the operators you'd expect: `+ - * /`. Two are worth calling out because
they're less common elsewhere:

- `//` — floor division (divide, then round down to the nearest whole number)
- `%`  — modulo (the remainder after division)
- `**` — exponentiation (power)

Also note: `/` always returns a `float`, even when dividing two whole numbers evenly.

```python
print(7 / 2)    # true division -> float
print(7 // 2)   # floor division -> int-like result
print(7 % 2)    # remainder
print(2 ** 5)   # 2 to the power of 5
```

## f-strings: building text from values

An f-string (`f"..."`) lets you embed variables directly inside a string using `{}`. This is
the standard, readable way to format output in modern Python — prefer it over manually
gluing strings together with `+`.

```python
name = "Grace"
score = 97

print(f"{name} scored {score} points")
```

## Recap

- Variables are names bound to values; Python infers the type. Use `snake_case`.
- Core types: `int`, `float`, `str`, `bool` (note `True`/`False` are capitalized).
- Arithmetic: `+ - * /` plus `//` (floor division), `%` (remainder), `**` (power). `/` always
  gives a `float`.
- Comparisons (`== != < > <= >=`) give `bool`; combine with `and`, `or`, `not`.
- f-strings (`f"...{value}..."`) are the standard way to build strings from variables.
