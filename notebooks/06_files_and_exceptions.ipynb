## Writing to a file

Use the built-in `open()` function with mode `"w"` (write) to create or overwrite a file.
The recommended pattern is a `with` statement — it automatically closes the file for you,
even if an error happens partway through, which is why it's preferred over calling
`.close()` manually.

```python
with open("data/notes.txt", "w") as f:
    f.write("Hello, file!\n")
    f.write("This is a second line.\n")

print("wrote data/notes.txt")
```

## Reading a file

Mode `"r"` (read, and the default if you omit the mode) opens a file for reading.
`.read()` returns the whole file as one string; looping over the file object directly gives
you one line at a time (each line includes its trailing `\n`).

```python
with open("data/notes.txt", "r") as f:
    contents = f.read()

print(contents)
```

```python
with open("data/notes.txt", "r") as f:
    for line in f:
        print(line.strip())  # .strip() removes the trailing newline
```

## Exceptions: things that go wrong

When something goes wrong — a file doesn't exist, a conversion fails — Python raises an
*exception*, and the program stops unless you handle it. Trying to open a missing file
raises `FileNotFoundError`; converting a non-numeric string with `int()` raises
`ValueError`.

```python
try:
    with open("data/does_not_exist.txt", "r") as f:
        contents = f.read()
except FileNotFoundError:
    print("That file doesn't exist yet!")
```

`try`/`except` lets you catch a specific exception type and handle it, instead of the whole
program crashing. Always name the exception type you expect (`except FileNotFoundError`)
rather than catching everything blindly — that way, unrelated bugs still surface as errors
instead of being silently swallowed.

```python
def to_int(text):
    try:
        return int(text)
    except ValueError:
        print(f"{text!r} isn't a valid number, using 0 instead")
        return 0

print(to_int("42"))
print(to_int("banana"))
```
