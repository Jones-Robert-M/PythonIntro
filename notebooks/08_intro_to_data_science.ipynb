## Why not just use lists?

You already know Python's `list`. `numpy` gives you an *array* type instead, which is much
faster for numeric work and lets you apply operations to every element at once ("vectorized"
operations) without writing a `for` loop.

````python
import sys


try:
    import numpy as np
except ImportError:
    print("Numpy is not installed.")
    # Now we will pip install numpy on the fly
    import subprocess
    subprocess.check_call([sys.executable, "-m", "pip", "install", "numpy"])
    import numpy as np

numbers = [1, 2, 3, 4, 5]      # a plain Python list
arr = np.array(numbers)         # a numpy array holding the same values

print(arr)
print(type(arr))
print(arr * 2)   # multiplies every element by 2, no loop needed
````

## A few numpy basics

Arrays have useful built-in aggregate methods: `.sum()`, `.mean()`, `.max()`, `.min()`. These
mirror what you'd otherwise write by hand with loops and the `statistics` module.

````python
scores = np.array([85, 92, 78, 90, 88])

print(scores.sum())
print(scores.mean())
print(scores.max())
````

## pandas: tables of data

A pandas `DataFrame` is a table — rows and columns, like a spreadsheet or a database table.
You typically build one from a dictionary mapping column names to lists of values.

````python
try:
    import pandas as pd
except ImportError:
    print("Pandas is not installed.")
    # Now we will pip install pandas on the fly
    import subprocess
    subprocess.check_call([sys.executable, "-m", "pip", "install", "pandas"])
    import pandas as pd

data = {
    "name": ["Ada", "Alan", "Grace"],
    "age": [36, 41, 85],
}
df = pd.DataFrame(data)
print(df)
````

Access a column like a dict key (`df["age"]`); it comes back as a pandas `Series`, which
behaves a lot like a numpy array — including supporting `.mean()`, `.max()`, etc.

````python
print(df["age"])
print(df["age"].mean())
````

## Filtering rows

You can filter a `DataFrame` using a boolean condition on a column, similar to the `if`
conditions you've written throughout this course — pandas applies the condition to every row
at once.

````python
df = pd.DataFrame({
    "name": ["Ada", "Alan", "Grace", "Linus"],
    "age": [36, 41, 85, 54],
})

adults_over_40 = df[df["age"] > 40]
print(adults_over_40)
````

## Recap
