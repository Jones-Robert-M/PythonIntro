## Importing modules

`import module_name` brings in a module's functions. Access them with `module_name.func()`.
You can also import specific names directly with `from module_name import name`.

```python
import math

print(math.sqrt(16))
print(math.pi)
```

```python
from math import sqrt, pi

print(sqrt(16))
print(pi)
```

## Useful standard library modules

A few modules you'll reach for often:

- `random` — random numbers and choices
- `datetime` — dates and times
- `os` — interacting with the operating system (files, paths — you saw this in
  [[06_files_and_exceptions]])
- `collections` — extra container types, e.g. `Counter` for counting things

```python
import random

print(random.randint(1, 6))          # simulate a dice roll
print(random.choice(["rock", "paper", "scissors"]))
```

```python
from datetime import date

today = date.today()
print(today)
print(today.year)
```

```python
from collections import Counter

votes = ["cat", "dog", "cat", "cat", "dog", "bird"]
print(Counter(votes))
```

## Writing your own modules

Any `.py` file is a module — you can `import` it from another file in the same folder. This
is how you organize larger programs: instead of one giant file, you split code into modules
by topic (e.g. a `math_helpers.py` with your own functions). Jupyter notebooks aren't
typically imported this way, but the functions you write in a notebook work the same as
functions in a `.py` module.

## Classes: bundling data and behavior

A `class` is a blueprint for creating objects that carry both data (attributes) and behavior
(methods). `__init__` is a special method that runs when you create a new object, setting up
its initial attributes. `self` refers to the specific object the method is being called on —
it's always the first parameter of a method.

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        return f"{self.name} says Woof!"

rex = Dog("Rex", 3)
print(rex.name)
print(rex.age)
print(rex.bark())
```
