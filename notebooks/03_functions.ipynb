## Defining and calling a function

Use `def` to define a function, followed by a name, parentheses for its parameters, and a
colon. The body is indented, same as `if`/`for`. `return` sends a value back to the caller —
without it, a function returns `None`.

Function names use `snake_case`, same as variables. A short docstring (a string literal right
after the `def` line) documents what the function does — it's convention to add one for any
function whose purpose isn't obvious from its name.

```python
def greet(name):
    """Return a greeting for the given name."""
    return f"Hello, {name}!"

print(greet("Marie"))
```

## Multiple parameters and default values

A function can take several parameters. You can give a parameter a default value, making it
optional when calling the function — callers can override it by passing their own value.

```python
def power(base, exponent=2):
    """Return base raised to exponent (default: square it)."""
    return base ** exponent

print(power(3))       # uses the default exponent of 2
print(power(3, 3))    # overrides it
```

## Returning multiple values

A function can return more than one value at once by separating them with commas — Python
packages them into a `tuple` automatically, and you can "unpack" them into separate variables
when calling the function.

```python
def min_and_max(numbers):
    """Return the smallest and largest values in numbers."""
    return min(numbers), max(numbers)

lowest, highest = min_and_max([4, 9, 1, 7])
print(lowest, highest)
```

## Why functions matter: avoiding repetition

A good sign you need a function is writing similar code more than once. Functions also make
code easier to test and reason about, since each one has a single, named responsibility —
this idea is sometimes summarized as "do one thing.

## Recap

- `def name(params):` defines a function; `return` sends back a value (`None` if omitted).
- Parameters can have default values, making them optional for the caller.
- A function can return multiple values as a tuple, which you can unpack directly.
- Extracting repeated logic into a well-named function makes code easier to read and reuse.
