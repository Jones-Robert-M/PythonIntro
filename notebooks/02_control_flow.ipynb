## `if` / `elif` / `else`

An `if` statement runs a block of code only when a condition is `True`. Python uses
indentation (4 spaces, by convention) to mark what's inside the block — there are no curly
braces. `elif` checks another condition if the first was false; `else` catches everything
else.

```python
temperature = 15

if temperature > 25:
    print("It's hot")
elif temperature > 10:
    print("It's mild")
else:
    print("It's cold")
```

## `for` loops

A `for` loop runs a block once per item in a sequence. The built-in `range(n)` produces the
numbers `0` up to (but not including) `n` — a common way to "do something n times".

```python
for i in range(5):
    print(i)
```

`for` loops also work directly over collections like lists, which you'll see more of in
the next notebook — for now, a quick preview:

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
```

## `while` loops

A `while` loop keeps running as long as its condition is `True`. Use it when you don't know
in advance how many times you'll need to repeat — for example, waiting for a value to reach
some threshold. Be careful: if the condition never becomes `False`, the loop never ends (an
"infinite loop").

```python
count = 0

while count < 3:
    print(f"count is {count}")
    count += 1  # without this line, the loop would never end
```

## `break` and `continue`

`break` exits a loop immediately. `continue` skips the rest of the current iteration and
moves to the next one. Both work in `for` and `while` loops.

```python
for number in range(10):
    if number == 5:
        break  # stop the loop entirely once we hit 5
    print(number)
```

```python
for number in range(10):
    if number % 2 == 0:
        continue  # skip even numbers, go straight to the next iteration
    print(number)
```

## Recap

- `if` / `elif` / `else` choose which block of code runs, based on conditions. Indentation
  defines the block.
- `for` loops repeat once per item in a sequence; `range(n)` gives `0..n-1`.
- `while` loops repeat as long as a condition holds — make sure the condition eventually
  becomes `False`.
- `break` exits a loop early; `continue` skips to the next iteration.
